## Output 1 (updated frequency calculator) Dayana

```python
def countfrequency(filename):
    file = open(filename)
    string=file.read()
    charset=[]
    for i in string:
        if i not in charset:
            charset.append(i)
    chardict={i:0 for i in charset}
    for i in string:
        for j in charset:
            if i==j:
                chardict[j]+=1
    n={}
    for i in chardict:
        n[i]=chardict[i]/len(string)
        if i==' ':
            n['space']=n.pop(' ')
        if i=='\n':
            n['newline']=n.pop('\n')
    return n
```
